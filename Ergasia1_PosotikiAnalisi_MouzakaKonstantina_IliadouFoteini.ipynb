### **Επιλέξτε ένα από τα κείμενα που είναι διαθέσιμα στο 75texts.zip και δημιουργήστε N (όπου Ν>30) αρχεία με ίσο (κατά το δυνατό) πλήθος λέξεων (με χρήση Python).**

```python
import numpy as np
import re
```

```python
init_text = ""
with open('75texts/Οι Ναΐτες πετάχτηκαν δίπλα - Γιώργος Τσαντίκος.pdf.v2.txt', encoding='utf8') as f:  
    init_text = f.read()


text = re.sub(r'[^\w\s]', ' ', init_text)
text = re.sub(r"\s+", " ", text)
text = text.lower()

words = text.split(" ")  
print(len(words))

words[-1]
```

```python
import os

i = 1
filename = "naites/n_" + str(i) + ".txt"
print("Creating file:", filename)
os.makedirs(os.path.dirname(filename), exist_ok=True)
with open(filename, 'w', encoding="UTF8") as output_file:
    for word in words:
        output_file.write(word + " ")
```

Χωρίσαμε το κείμενό μας σε μικρότερα κείμενα μήκους των 300 λέξεων, καθώς βάσει εμπειρικού κανόνα ένα κείμενο 300 λέξεων έχει τα χαρακτηριστικά που θέλουμε και τα αποτελέσματα είναι πιο ασφαλή. 

```python
limit = 300
total_files = round(len(words)/limit)

if total_files*limit<len(words):
    total_files =  total_files + 1
print("number of files\t", total_files)

for i in np.arange(1, total_files+1):
    start = (i-1)*limit
    end = min(i*limit, len(words))
    temp = words[start:end]
    filename = "naites/n_" + str(i) + ".txt"
    print("Creating file:", filename)
    output_file = open(filename, 'w', encoding ="UTF8")
    for word in temp:
        output_file.write(word + " ")
    output_file.close()
```

### **Yπολογίστε τη μέση, τη διάμεση τιμή και την τυπική απόκλιση για κάθε δείκτη.**

### **Σχολιάστε την κατανομή των τιμών για κάθε δείκτη (π.χ. ομοιότητα με την κανονική κατανομή, skewness/ασυμμετρία).**

#### **Skewness**
Μετράει το βαθμό της συμμετρίας των δεδομένων ως προς τη συχνότητα. Μας δείχνει αν η κατανομή μας είναι συμμετρική ή όχι. Στην ουσία πρόκειται για μία σύγκριση της κατανομής μας με την κανονική κατανομή.

```python
from scipy.stats import skew
skewness = skew(texts_results['TTR'])
print("\nskewness:\t", skewness)

if skewness < 0:
    print("****left-skewed distribution (negatively skewed)****\n" )
if skewness > 0:
    print("****right-skewed distribution (positively skewed)****\n" )
hist = TTR_column_2.hist(bins=16, density=True)
sns.kdeplot(TTR_column_2)
plt.show()
```

Βάσει της τιμής ασυμμετρίας '1.43', η κατανομή του TTR είναι πολύ στρεβλή. Αυτό διότι τιμές μεγαλύτερες του 1 συνεπάγονται στρεβλή κατανομή (1.43 > 1). Πρόκειται για μία θετικά ασύμμετρη κατανομή, καθώς στο διάγραμμα η δεξιά ουρά είναι μεγαλύτερη από την αριστερή.


```python
skewness = skew(texts_results['h-Point'])
print("\nskewness:\t", skewness)

if skewness < 0:
    print("****left-skewed distribution (negatively skewed)****\n" )
if skewness > 0:
    print("****right-skewed distribution (positively skewed)****\n" )
hist = H_column_2.hist(bins=16, density=True)
sns.kdeplot(H_column_2)
plt.show()
```

#### **Kurtosis**
Η Κύρτωση μας λέει αν μία κατανομή είναι ψηλότερη ή κοντύτερη από την κανονική κατανομή. Μετράει το βαθμό συγκέντρωσης των δεδομένων γύρω από τη μέση τιμή. Δείχνει την αιχμηρότητα ή την πλάτυνση της κατανομής.

```python
from scipy.stats import kurtosis
TTR_kurtosis = kurtosis(texts_results['TTR'])
print("\nkurtosis:\t", TTR_kurtosis)

thr = 4* (6/res.get('count'))**0.5
print("rule of thumb threshold:\t", thr)
if abs(TTR_kurtosis) > thr:
    if TTR_kurtosis > 0:
        print("****leptokurtic****\n" )
    if TTR_kurtosis < 0:
        print("****platykurtic****\n" )
else:
    print("****mesokurtic****\n" )
hist = TTR_column_2.hist(bins=16, density=True)
sns.kdeplot(TTR_column_2)
plt.show()
```

Η κατανομή του TTR είναι λεπτόκυρτη, καθώς παρουσιάζει μεγάλη συγκέντρωση τιμών γύρω από την περιοχή της μέσης τιμής. Η κατανομή είναι ψηλότερη από την κανονική κατανομή.
